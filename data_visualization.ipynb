## Diffencies between Matplotlib and Seaborn

Matplotlib and Seaborn are both Python visualization libraries that are widely used for creating high-quality graphs, charts, and other visualizations. While they share some similarities, there are several key differences between the two libraries:

- Default Aesthetics: Seaborn is built on top of Matplotlib and provides a higher-level interface for creating statistical graphics with a more attractive default aesthetic. This means that Seaborn is designed to create aesthetically pleasing visualizations with minimal customization, while Matplotlib provides a more basic interface that requires more customization to achieve an attractive look.

- Type of Plots: While both libraries can create a wide range of plot types, Seaborn is particularly strong in creating statistical visualizations such as heatmaps, pair plots, and distribution plots, while Matplotlib has a wider range of plot types and is more suitable for creating custom visualizations that require precise control over the appearance of each element.

- Complexity: Seaborn offers a simplified API that requires fewer lines of code to create complex visualizations, making it ideal for users who want to create high-quality graphics with less programming effort. Matplotlib offers a more granular API that allows for a greater degree of customization, but can be more complex to use.

- Flexibility: While Seaborn offers a simplified API and a limited range of plot types, Matplotlib is a more flexible and low-level library that allows for the creation of custom visualizations with a greater degree of control over every element of the plot.

- Integration with Pandas: Seaborn is tightly integrated with Pandas, a popular data analysis library in Python, making it easy to use Seaborn to visualize data stored in a Pandas DataFrame. Matplotlib can also be used with Pandas, but requires more manual data manipulation and plotting code.

Overall, Seaborn is a good choice for creating attractive statistical visualizations with minimal customization, while Matplotlib is a more flexible library that allows for greater control over the appearance of the final plot.

## Matplotlib

````python
x = np.linspace(0,10,100) #create 100 values ​​between 0 and 10
y = np.sin(x)
z = np.cos(x)

# sine wave
plt.plot(x,y, 'xr')
plt.show() #show graph
````

As you can see in the code above, we create a 100 points using numpy's `linspace` function. Then, we create a second array of 100 points using the `sin` function. Finally, we plot the two arrays using the `plot` function, also make the visualization with red x's.

But this chart needs some improvements. We can add a title, labels for the x and y axis, and a legend.

````python
# adding title, x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
````

Matplotlib have a lot of different types of charts, like bar charts, pie charts, scatter plots, etc. In the example below, we will create a bar chart.

````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) #size chart
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people) #create chart
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
````

Now a example for a pie chart:

````python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%') #number of decimals
plt.show()
````

## Seaborn

````python
# Load the example dataset 
tips = sns.load_dataset('tips')
# setting a theme for the plots
sns.set_theme()
````

````python
# visualize the data
sns.relplot(data=tips, x ='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
````

In the example above, we analyzed a dataset about the difference between hour of the day, smoker zone and size of the table and their relation with the total_bill and tip.

As you can see in this example there is more data displayed in the same chart, and the chart is more complex. This is because we are using the Seaborn library.

````python
# loading the titanic dataset
titanic = sns.load_dataset('titanic')
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
````

As you can see in the example above we can display also different types of charts using Seaborn library. In the example above we used a bar chart for display the rate of the survived passengers of the Titanic, based on their sex and the class they were travelling, but we can also use a pie chart, scatter plot, etc.
