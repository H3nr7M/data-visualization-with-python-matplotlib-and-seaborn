#Data_visualization
In this notebook we will address one of the most important tasks in data processing, the visualization

```python
# importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,100) #create 100 values ​​between 0 and 10
y = np.sin(x)
z = np.cos(x)
# sine wave
plt.plot(x,y, 'xr')
plt.show() #show graph
```

```python
# adding title, x-axis & y-axis labels
plt.plot(x,y, 
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) #size chart
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people) #create chart
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%') #number of decimals
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
# setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x ='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
# loading the titanic dataset
titanic = sns.load_dataset('titanic')
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```
